```python
import requests
from bs4 import BeautifulSoup

# URL inicial
url = "https://dev.mysql.com/downloads/connector/j/"

# Crear una sesión
session = requests.Session()

# Obtener la página inicial
response = session.get(url)
if response.status_code != 200:
    print("Error al acceder a la página.")
    exit()
```

```python
# Parsear la página
soup = BeautifulSoup(response.content, "html.parser")
```

```python
import asyncio
from playwright.async_api import async_playwright

async def download_mysql_connector():
    async with async_playwright() as p:
        # Lanzar un navegador en modo headless
        browser = await p.firefox.launch(headless=True)  # También puedes usar 'chromium' o 'webkit'
        page = await browser.new_page()

        # Abrir la página
        await page.goto("https://dev.mysql.com/downloads/connector/j/")

        # Seleccionar "Ubuntu Linux" en el menú desplegable
        await page.select_option("select#os", "Ubuntu Linux")  # Cambia el selector si es necesario

        # Seleccionar "Ubuntu Linux 22.04" en el segundo menú desplegable
        await page.select_option("select#os_version", "Ubuntu Linux 22.04 (Architecture Independent)")

        # Hacer clic en el botón de descarga
        async with page.expect_download() as download_info:
            await page.click("text=Download")  # Busca el botón con texto "Download"
        download = await download_info.value

        # Guardar el archivo descargado
        await download.save_as("mysql-connector.deb")
        print("Archivo descargado correctamente como 'mysql-connector.deb'.")

        # Cerrar el navegador
        await browser.close()

# Ejecutar la función asincrónica
await download_mysql_connector()
```
